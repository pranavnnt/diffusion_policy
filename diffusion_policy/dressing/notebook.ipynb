```python
import zarr, numpy as np, cv2, matplotlib.pyplot as plt

def preprocess_zarr(
    src_path, dst_path, 
    crop_box=(0, -200, 100, -100),   # (h_start, h_end, w_start, w_end)
    resize_shape=(128, 128),         # (h, w)
    step=256,                        # frames per chunk to process
    show_every=2000                  # display every Nth frame
):
    src = zarr.open(src_path, "r")
    dst = zarr.open(dst_path, "w")

    dst_data = dst.create_group("data")
    dst_meta = dst.create_group("meta")

    for k, arr in src["data"].items():
        if k == "image":
            T, C, H, W = arr.shape
            out_h, out_w = resize_shape
            shape = (T, C, out_h, out_w)
            chunks = (arr.chunks[0], C, out_h, out_w)

            dst_arr = dst_data.create_dataset(
                name=k, shape=shape, chunks=chunks,
                dtype="uint8", compressor=arr.compressor
            )

            for i in range(0, T, step):
                print(f"Processing frames {i}:{i+step}")
                batch = arr[i:i+step]  # (step, 3, H, W)
                resized = np.empty((batch.shape[0], C, out_h, out_w), dtype=np.uint8)

                for j in range(batch.shape[0]):
                    img = np.transpose(batch[j], (1, 2, 0))  # (H, W, C), RGB

                    # crop
                    hs, he, ws, we = crop_box
                    img = img[hs:he, ws:we]

                    # resize to target (default 128x128)
                    img = cv2.resize(img, resize_shape, interpolation=cv2.INTER_AREA)
                    
                    # scale to [0,255] and cast to uint8
                    img = np.clip(img * 255.0, 0, 255).astype(np.uint8)
                     # back to (C, H, W)
                    resized[j] = np.transpose(img, (2, 0, 1))

                    # show preview
                    global_idx = i + j
                    if global_idx % show_every == 0:
                        plt.imshow(img)   # already RGB, no cv2.cvtColor
                        plt.title(f"Frame {global_idx}")
                        plt.axis("off")
                        plt.show()

                dst_arr[i:i+step] = resized

        else:
            # copy other arrays directly
            dst_data.create_dataset(
                name=k, data=arr[:],
                chunks=arr.chunks, compressor=arr.compressor
            )

    # copy metadata
    for k, arr in src["meta"].items():
        dst_meta.create_dataset(name=k, data=arr[:])

    print(f" Finished preprocessing. Saved to {dst_path}")


# Example usage in notebook
preprocess_zarr(
    src_path="/home/pnt8/workspace/dressing_sim_ws/diffusion_policy/diffusion_policy/data/first_arm_scene/seed0_combined.zarr",
    dst_path="/home/pnt8/workspace/dressing_sim_ws/diffusion_policy/diffusion_policy/data/first_arm_scene/seed0_combined_compressed.zarr",
    crop_box=(50, -50, 50, -50),   # crop before resize
    resize_shape=(128, 128),       # downsample to 128x128
    step=128,
    show_every=500
)
```

```python
# create empty destination arrays with same shape, but uint8
for k, arr in src["data"].items():
    if k == "image":
        shape = arr.shape
        chunks = arr.chunks
        compressor = arr.compressor
        dst_arr = dst.create_dataset(f"data/{k}",
                                     shape=shape,
                                     chunks=chunks,
                                     dtype='uint8',
                                     compressor=compressor)
        # process in slices along time axis
        step = 256  # tune this to your RAM
        for i in range(0, shape[0], step):
            print(f"Converting {i}:{i+step}")
            batch = arr[i:i+step]        # load small chunk
            dst_arr[i:i+step] = batch.astype(np.uint8)
    else:
        dst.create_dataset(f"data/{k}",
                           data=arr[:],
                           chunks=arr.chunks,
                           compressor=arr.compressor)

dst.create_group("meta")
dst["meta/episode_ends"] = src["meta/episode_ends"][:]
```
